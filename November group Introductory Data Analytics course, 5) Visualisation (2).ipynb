```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
os.getcwd()
Olympics= pd.read_csv("C:\\Users\\Jim/Downloads/athlete_events.csv/athlete_events.csv")
swim= Olympics[Olympics['Sport'] == 'Swimming']
Medals=['Gold', 'Silver', 'Bronze']; winner=swim['Medal'].isin(Medals)
swim_champ=pd.DataFrame(swim[winner])
swim_champ_year=swim_champ.sort_values('Year', ascending = False)
modern_champ= swim_champ_year[swim_champ_year['Year'] > 1990]
gold=swim_champ[swim_champ['Medal']== 'Gold']
```

```python
# GENERATING CHARTS
```

```python
# LINE PLOT
```

```python
#After importing seaborn and matplotlib, I begin with a simple chart showing age of gold winning athletes over the years.
#As Year is ordinal data and Age is continuous, I opt for a lineplot

age=np.array(gold['Age'])
year=np.array(gold['Year'])
sns.lineplot(year, age, color='orange').set(title="Gold medal winner ages over the years")
plt.xlabel("Year")
plt.ylabel("Age")
plt.show()
```

```python
#From the plot, we see that athlete ages dropped from 1940, spiked in the 1950's and rose steadily in the 1980's to higher than ever in the 2010's
# Modern athletes are probably older due to advances in sports and medical science extending longevity
#The shadow around the line plot represents the confidence interval
```

```python
# COUNT PLOT
```

```python
#The next chart is a count plot showing the countries that win the most gold medals, as this is categorical data
#with the hue command used to show the proportion of male and female winners from each country
#I begin by trying a simple line plot with the gold dataset
all_time=sns.countplot(x='Team', data=gold, hue='Sex')
```

```python
#This is a mess, we cannot distinguish country names on the x axis
#So we set the x tick labels and then rotate them by 90 degrees son that the names can be read
all_time=sns.countplot(x='Team', data=gold, hue='Sex')
all_time.set_xticklabels(all_time.get_xticklabels(), rotation=90)
```

```python
#This is better, but the volume of countries makes it hard to read their names
#So we opt for two count plots, one for the 21st century, and one for games from 1896 to 1920
new_millenium= gold[gold['Year'] > 2000]
```

```python
modern_winning_countries=sns.countplot(x='Team', data=new_millenium, hue='Sex')
modern_winning_countries.set_xticklabels(modern_winning_countries.get_xticklabels(), rotation=90)
modern_winning_countries.set(title= "Gold winning countries of the 21st century")
```

```python
#This looks much clearer, and agrees with the html data retrieved in Chapter 2 of the project on the top performing countries in Olympic swimming
#United states and Australia dominate the swimming gold medal winning countries
#United States male swimmers win more than their female counterparts (on average)
#And female Australian swimmers win more medals for their country than males do (on average)
# and now for the plot of past gold winning countries
history=gold[gold['Year'] < 1910]
```

```python
past_winning_countries=sns.countplot(x='Team', data=history, hue='Sex' )

past_winning_countries.set_xticklabels(past_winning_countries.get_xticklabels(), rotation=90)
past_winning_countries.set(title= "Gold winning countries between 1896 and 1920")
```

```python
#In these years, Olympic Swimming was dominated by Great Britain and Hungary
#We also see no female medal winners as women were not allowed to compete in the games in those times
```

```python
#For an extra barplot, I plotted gold medal winning male swimmer's age against different events in Olympic swimming
#I can do this using the groupby() method seen earlier

men_swim_gold=gold[gold['Sex']== 'M']

men_swim_gold.groupby(['Event'])['Age'].mean().plot.bar(legend=True).set(title="Male swimmer ages in different events")

plt.ylabel("Age in years")
plt.show()
```

```python
#Interestingly, I found that older athletes usually do well in events over longer distances requiring greater endurance
#This indicates that younger swimmers are better suited to short-distance events where greater speed and power are needed
```

```python
#SCATTER PLOT
```

```python
#My final plot is a simple scatter plot showing the relationship between bodyweight and age in gold medal winning swimmers
#Both are categorical data
weight=np.array(gold['Weight'])
sns.scatterplot(x=age, y=weight).set(title="Relationship between weight and age in gold winning swimmers")
plt.xlabel("Age in years")
plt.ylabel("Weight in kilograms")
plt.show()
```

```python
#Here, we see that genearlly, older athletes tend to be heavier, with the heaviest athletes being in their twenties
```
